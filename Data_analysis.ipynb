```python
# Mapping dictionaries
gender_map = {1: 'Male', 2: 'Female'}
married_map = {0: 'Single', 1: 'Married'}
dependents_map = {0: '0', 1: '1', 2: '2', 3: '3+'}
graduate_map = {0: 'No', 1: 'Yes'}
self_employed_map = {0: 'No', 1: 'Yes'}
credit_history_map = {0: 'No', 1: 'Yes'}
property_area_map = {1: 'Urban', 2: 'Semiurban', 3: 'Rural'}

# Apply the mappings to both datasets
df['Gender'] = df['Gender'].replace(gender_map)
df['Married'] = df['Married'].replace(married_map)
df['Dependents'] = df['Dependents'].replace(dependents_map)
df['Graduate'] = df['Graduate'].replace(graduate_map)
df['Self_Employed'] = df['Self_Employed'].replace(self_employed_map)
df['Credit_History'] = df['Credit_History'].replace(credit_history_map)
df['Property_Area'] = df['Property_Area'].replace(property_area_map)


print("Dataset  (Mapped):")
display(df.head())
```

### Outlier analysis

```python


numeric_columns = [col for col in df.columns if df[col].dtype != 'object']
num_plots = len(numeric_columns)

# Create a figure with subplots
fig, axes = plt.subplots(nrows=(num_plots + 2) // 3, ncols=3, figsize=(15, (num_plots + 2) // 3 * 5))
axes = axes.flatten()

# Plot each numeric column in a separate subplot
for i, column in enumerate(numeric_columns):
    sns.boxplot(y=df[column], ax=axes[i])
    axes[i].set_title(f'Box Plot of {column}')
    axes[i].set_xlabel('')

# Hide any unused subplots
for j in range(len(numeric_columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python

def remove_outliers(df):

    df_clean = df.copy()


    for column in df_clean.select_dtypes(include=['number']).columns:
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df_clean[column].quantile(0.25)
        Q3 = df_clean[column].quantile(0.75)
        IQR = Q3 - Q1

        # Define bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter out outliers
        df_clean = df_clean[(df_clean[column] >= lower_bound) & (df_clean[column] <= upper_bound)]

    return df_clean

# Remove outliers from the DataFrame
df_no_outliers = remove_outliers(df)

# Verify the shape of the original and cleaned DataFrame
print(f"Original DataFrame shape: {df.shape}")
print(f"DataFrame shape after removing outliers: {df_no_outliers.shape}")
```
